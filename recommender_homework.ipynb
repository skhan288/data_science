```python
import pandas as pd
import numpy as np
```

```python
M = pd.read_csv("../../Movie Reviews.csv", index_col='Name')
```

```python
def pearson(s1, s2):
    """Take two pd.Series objects and return a pearson correlation."""
    s1_c = s1 - s1.mean()
    s2_c = s2 - s2.mean()
    return np.sum(s1_c * s2_c) / np.sqrt(np.sum(s1_c ** 2) * np.sum(s2_c ** 2))
```

```python
M
```

```python
def get_recs(movie_name, M, num):

    import numpy as np
    reviews = []
    for title in M.columns:
        if title == movie_name:
            continue
        cor = pearson(M[movie_name], M[title])
        if np.isnan(cor):
            continue
        else:
            reviews.append((title, cor))
    
    reviews.sort(key=lambda tup: tup[1], reverse=True)
    return reviews[:num]

    
```

```python
def pred_rate(x, a):

    j=0
    k=0
    for i in range(0, x.shape[0], 1): # for each row in x (range(0,23))
        if x.iloc[i]>0 and a[i]>0: 
            j= j+(a[i]*x.iloc[i])
            k= k+a[i]
        else:
            continue
    return j/k
```
